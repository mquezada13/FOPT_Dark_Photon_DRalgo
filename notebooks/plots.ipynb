```python
import numpy as np
import matplotlib.pyplot as plt

# --- helper to sort x-values
def sort_xy(a):
    a = np.asarray(a)
    idx = np.argsort(a[:,0])
    return a[idx]

# --- helper to safely fill between two (x,y) curves ---
def fill_between_curves(ax, A, B, color='royalblue', ngrid=None):
    """
    Fill the area between two (x,y) curves, even if x-grids differ.
    """
    A = np.asarray(A)
    B = np.asarray(B)
    A = A[np.argsort(A[:,0])]
    B = B[np.argsort(B[:,0])]

    xmin = max(A[0,0], B[0,0])
    xmax = min(A[-1,0], B[-1,0])
    if not (xmax > xmin):
        print("Warning: no overlapping x-range between curves.")
        return

    # choose grid: denser one or uniform linspace
    if ngrid is None:
        mA = (A[:,0] >= xmin) & (A[:,0] <= xmax)
        mB = (B[:,0] >= xmin) & (B[:,0] <= xmax)
        xA = A[mA, 0]
        xB = B[mB, 0]
        x = xA if xA.size >= xB.size else xB
    else:
        x = np.linspace(xmin, xmax, int(ngrid))

    yA = np.interp(x, A[:,0], A[:,1])
    yB = np.interp(x, B[:,0], B[:,1])

    m = np.isfinite(x) & np.isfinite(yA) & np.isfinite(yB)
    if np.any(m):
        ax.fill_between(x[m], yA[m], yB[m], color=color, alpha=0.15, lw = 0, label =r"$q^\ast =17-40\, \mathrm{MeV} $")
    else:
        print("Warning: no valid finite values to fill.")

# --- ordenar por x
file1 = sort_xy(file1)
file2 = sort_xy(file2)
file3 = sort_xy(file3)
file4 = sort_xy(file4)
file5 = sort_xy(file5)

# --- desempaquetar
x1, y1 = file1[:,0], file1[:,1]
x2, y2 = file2[:,0], file2[:,1]
x3, y3 = file3[:,0], file3[:,1]
x4, y4 = file4[:,0], file4[:,1]
x5, y5 = file5[:,0], file5[:,1]

# --- create figure
fig, ax = new_figure(figsize=(7, 5.5), dpi=120)
fig.patch.set_facecolor('white')

# --- fill between file5 and file1
fill_between_curves(ax, file5, file1, color='limegreen')

# --- plot curves
#ax.plot(x1, y1, color='orange', lw=1.5)
ax.plot(x2, y2, color='limegreen', lw=2, ls=':', label =r"$q^\ast =20\, \mathrm{MeV} $")
ax.plot(x3, y3, color='limegreen', lw=1.0,ls = "--", label =r"$q^\ast =25\, \mathrm{MeV} $")
ax.plot(x4, y4, color='limegreen', lw=1.0, ls = "-", label =r"$q^\ast =30\, \mathrm{MeV} $")
#ax.plot(x5, y5, color='gray', lw=1.5)
ax.text(0.7, 0.32, r"$q_{\rm min} = 1.7\,\rm MeV $", fontsize=13, ha='left', va='bottom')
ax.text(0.7, 0.3, r"$m_\gamma = 1\,\rm MeV $", fontsize=13, ha='left', va='bottom')


# --- formatting
ps.apply_standard_formatting(
    ax,
    ylabel=r"$F_{\mathrm{soft}}(\nu)$",
    xlabel=r"$\nu$",
    xlim=(0, 0.95),
    ylim=(0, 0.35),
    xlog=False,
    ylog=False,
)

plt.style.use('classic')
plt.legend(loc="best", ncol=1, fontsize=13, frameon=False)
plt.tight_layout()
plt.savefig("/Users/ramirezmaura/Library/CloudStorage/Dropbox/Maura-MaElena/Articulo-Tesis/data/Fsoft_plot.pdf",
            dpi=300, bbox_inches="tight", facecolor="white")
```
